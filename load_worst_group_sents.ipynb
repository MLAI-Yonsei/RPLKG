```python
import re
import requests
import pandas as pd
```

```python
words = ['collie', 'canoe', 'tights', 'missile', 'sunglasses']
```

```python
word_list = []
sent_list = []
for word in words:
    len_response = 1
    i = 0
    while len_response > 0:
        response = requests.get(f'http://api.conceptnet.io/c/en/{word}?offset={i * 1000}&limit=1000').json()
        edges = response['edges']
        len_response = len(edges)
        i += 1
        for edge in edges:
            surface_text = edge['surfaceText']
            if surface_text is None:
                continue
            surface_text = surface_text.replace('[', '').replace(']', '')
            a = re.match('^[a-zA-Z0-9 ]*$', surface_text)
            if a is None:
                continue
            print(word, f'{surface_text}. a photho of a {word}.')
            p_sent = f'{surface_text}. a photho of a {word}.'
            word_list.append(word)
            sent_list.append(p_sent)
```

```python
df_dict = {'word': word_list,
           'sent': sent_list}

df = pd.DataFrame(df_dict)
df
```

```python
df = df.dropna()
df
```

```python
df.to_csv('worst_sents.csv', index=False, encoding='utf-8')
```
